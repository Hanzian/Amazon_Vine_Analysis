```python
# Import 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read csv files
from google.colab import files

uploaded = files.upload()

import io
df = pd.read_csv(io.StringIO(uploaded['vine_table.csv'].decode('utf-8')))
```

```python
vine_df= df
vine_df.head()
```

```python
vine_df.info()
```

# **Filter by total votes**

```python
filtered_vine_df = vine_df[vine_df['total_votes'] >= 20]
filtered_vine_df.head()
```

# **Find most helpful reviews**


```python
help_vine_df = filtered_vine_df.loc[filtered_vine_df.helpful_votes / filtered_vine_df.total_votes >= 0.5]
help_vine_df.head()
```

# **Create a DataFrame or table where there is a Vine review**

```python
vine_reviews = help_vine_df[help_vine_df['vine'] == 'Y']
vine_reviews.head()
```

# **Create a DataFrame or table where there isn’t a Vine review**

```python
no_vine_reviews = help_vine_df[help_vine_df['vine'] == 'N']
no_vine_reviews.head()
```

# **5 Stars reviews Analysis**

## **The total number of reviews**

Retrieve the count of vine and no_vine reviews

```python
count_vine = len(vine_reviews)
count_vine
```

```python
count_no_vine = len(no_vine_reviews)
count_no_vine
```

```python
total_vines= count_vine + count_no_vine
total_vines
```

```python

max_rating_vine = len(vine_reviews[vine_reviews['star_rating']==5])
max_rating_vine
```

```python
max_rating_nonvine = len(no_vine_reviews[no_vine_reviews['star_rating']==5])
max_rating_nonvine
```

```python
# 5-Star Review percentage by program
star_vine = max_rating_vine/count_vine
star_vine
```

```python
star_no_vine = max_rating_nonvine/count_no_vine
star_no_vine
```
